# Case 1: Telecom — Identifying Inefficient Operators

## Business Context
The virtual phone service **CallMeMaybe** needs a tool that provides supervisors with information about underperforming operators. The inefficiency criteria are:
- High rate of **missed incoming calls** (internal and external)
- **Long wait times** on incoming calls
- **Low number of outgoing calls** (if the operator has outbound duties)

## Available Data
| Dataset | Rows | Description |
|---------|------|-------------|
| `telecom_dataset_new.csv` | 53,902 | Call records (direction, duration, operator, etc.) |
| `telecom_clients.csv` | 732 | Client information (tariff plan, registration date) |

### Initial Data Observations
- `operator_id` has **8,172 null values** (~15%) → investigate if these are calls without an assigned operator
- `internal` has **117 nulls** → determine if these are errors or a third call type
- The `date` column is string type → convert to datetime
- `total_call_duration` includes wait time; `call_duration` does not → the difference is the **wait time**

## Work Plan — Case 1

### Step 1: Data Loading and Initial Exploration
- [ ] Import libraries (pandas, numpy, matplotlib, seaborn, scipy)
- [ ] Load both datasets (`telecom_dataset_new.csv` and `telecom_clients.csv`)
- [ ] Examine structure: `.info()`, `.describe()`, `.head()`, `.shape`
- [ ] Identify incorrect data types
- [ ] Quantify null values and duplicates

### Step 2: Preprocessing and Data Cleaning
- [ ] Convert `date` to datetime format
- [ ] Convert `date_start` in the clients dataset to datetime
- [ ] Evaluate and handle null values in `operator_id` (are these calls without an operator? remove or impute?)
- [ ] Evaluate and handle nulls in `internal`
- [ ] Find and remove duplicates if they exist
- [ ] Verify consistency: `call_duration` ≤ `total_call_duration`
- [ ] Create derived column: `wait_time` = `total_call_duration` - `call_duration`

### Step 3: Exploratory Data Analysis (EDA)
- [ ] **General call distribution:** histograms of duration, wait time, call count
- [ ] **Analysis by direction** (`in` vs `out`): volume, average duration, missed call rate
- [ ] **Analysis by type** (internal vs external): usage patterns
- [ ] **Temporal analysis:** daily/weekly call trends
- [ ] **Operator analysis:** key metrics per operator (missed calls, wait times, outgoing calls)
- [ ] **Client analysis:** distribution by tariff, tenure, relationship with call volume
- [ ] Merge datasets by `user_id` to enrich the analysis

### Step 4: Identifying Inefficient Operators
- [ ] Define quantitative inefficiency metrics per operator:
  - Missed incoming call rate (internal + external)
  - Average wait time on incoming calls
  - Total outgoing calls (for outbound operators)
- [ ] Establish thresholds to classify operators as "inefficient" (using percentiles, z-scores, or IQR criteria)
- [ ] Generate operator ranking by inefficiency level
- [ ] Visualize inefficient vs. efficient operators

### Step 5: Statistical Hypothesis Testing
- [ ] **Hypothesis 1:** Is the missed call rate of inefficient operators significantly higher than the rest? (t-test or Mann-Whitney U)
- [ ] **Hypothesis 2:** Does the average wait time differ significantly between inefficient and efficient operators?
- [ ] **Hypothesis 3:** Are there significant differences in operator metrics based on the client's tariff plan?
- [ ] Define significance level (α = 0.05)
- [ ] Check normality assumptions (Shapiro-Wilk) and choose parametric or non-parametric tests
- [ ] Document null hypothesis (H₀) and alternative hypothesis (H₁) for each test

### Step 6: Conclusions and Recommendations
- [ ] Summarize key EDA findings
- [ ] List operators identified as inefficient with quantitative evidence
- [ ] Formulate actionable recommendations for supervisors
- [ ] Document analysis limitations

### Step 7: Dashboard (Tableau)
- [ ] Call duration histogram
- [ ] Pie chart: internal vs external calls
- [ ] Interactive filters by call direction and type (internal/external)
- [ ] Histogram of calls per day
- [ ] Publish on Tableau Public

### Step 8: PDF Presentation
- [ ] Create presentation with key conclusions
- [ ] Include relevant visualizations
- [ ] List 5-10 sources consulted with brief descriptions
- [ ] Export to PDF

# Case 2: A/B Test — Recommender System Validation

## Business Context
An international online store launched an A/B test to evaluate an **improved recommender system**. Group B (new payment funnel) is expected to show at least a **10% improvement in conversion** at each funnel stage:

`product_page → product_cart → purchase`

## Technical Specifications
| Parameter | Value |
|-----------|-------|
| Test name | `recommender_system_test` |
| Groups | A (control) vs B (new funnel) |
| Launch date | 2020-12-07 |
| Registration cutoff | 2020-12-21 |
| End date | 2021-01-01 |
| Audience | 15% of new EU region users |
| Expected participants | 6,000 |

## Available Data
| Dataset | Rows | Description |
|---------|------|-------------|
| `final_ab_events_upd_us.csv` | 423,761 | User events (login, product_page, product_cart, purchase) |
| `final_ab_new_users_upd_us.csv` | 58,703 | Users registered Dec 7-21, 2020 |
| `final_ab_participants_upd_us.csv` | 14,525 | Test participants (A: 8,214 / B: 6,311) |
| `ab_project_marketing_events_us.csv` | 14 | 2020 marketing events calendar |

### Initial Observations
- There are **2 tests** in the participants dataset: `recommender_system_test` and `interface_eu_test` → filter only the first one
- Groups are unbalanced (A: 8,214 vs B: 6,311) → verify if this is a problem
- A **"Christmas & New Year Promo"** marketing event overlaps with the test (Dec 25 - Jan 3) → potential contamination
- The `product_cart` event in the data corresponds to `product_card` mentioned in the technical specs
- `details` has ~363K nulls → this is expected, only filled for `purchase` events

## Work Plan — Case 2

### Step 1: Define Study Objectives
- [ ] Clearly document the A/B test objectives and success criteria
- [ ] Define the expected conversion funnel and improvement thresholds (≥10% per stage)

### Step 2: Data Loading and Exploration
- [ ] Load all 4 datasets
- [ ] Examine structure, data types, nulls, and duplicates in each
- [ ] Convert date columns to datetime (`event_dt`, `first_date`, `start_dt`, `finish_dt`)
- [ ] Handle missing values and duplicates as appropriate

### Step 3: Test Configuration Validation
- [ ] Filter only participants from `recommender_system_test`
- [ ] Verify **no users appear in both groups** (A and B simultaneously)
- [ ] Verify no users are in both tests simultaneously
- [ ] Check that the audience corresponds to ~15% of new EU users
- [ ] Assess whether the participant count approaches the expected 6,000
- [ ] Verify test users registered between December 7 and 21
- [ ] Identify **marketing events overlapping** with the test period (Christmas & New Year Promo)

### Step 4: Exploratory Data Analysis (EDA)
- [ ] Event distribution by day → visualize temporal trends
- [ ] Event count distribution per user → is it equitable between A and B?
- [ ] Conversion funnel analysis by group:
  - login → product_page → product_cart → purchase
  - Calculate conversion rate at each stage
- [ ] Visualize comparative funnel (A vs B)
- [ ] Identify data peculiarities (outliers, anomalous behaviors, Christmas event effect)

### Step 5: Statistical Evaluation of the A/B Test
- [ ] Calculate conversion proportions by group at each funnel stage
- [ ] Apply **z-test for proportions** at each stage:
  - product_page: group A vs group B
  - product_cart: group A vs group B
  - purchase: group A vs group B
- [ ] Define H₀ (no difference) and H₁ (group B has higher conversion) for each test
- [ ] Significance level: α = 0.05 (apply Bonferroni correction if needed: α/3)
- [ ] Interpret p-values and determine if the expected 10% improvement is achieved

### Step 6: Conclusions
- [ ] Summarize EDA findings (issues identified in test execution)
- [ ] Interpret statistical results
- [ ] Issue a verdict: Should the new recommender system be implemented?
- [ ] Document limitations and recommendations for future tests

# Case 3: SQL — Book Database Analysis

## Business Context
A digital reading service wants to generate a value proposition for a new product, using data about books, publishers, authors, and user ratings.

## Database Structure
| Table | Description | Key Columns |
|-------|-------------|-------------|
| `books` | Book data | book_id, author_id, title, num_pages, publication_date, publisher_id |
| `authors` | Author data | author_id, author |
| `publishers` | Publisher data | publisher_id, publisher |
| `ratings` | User ratings | rating_id, book_id, username, rating |
| `reviews` | Text reviews | review_id, book_id, username, text |

## Required Queries
1. Number of books published after January 1, 2000
2. Number of reviews and average rating per book
3. Publisher with the most books over 50 pages
4. Author with the highest average rating (only books with ≥50 ratings)
5. Average number of text reviews among users who rated 50+ books

## Work Plan — Case 3

### Step 1: Setup and Connection
- [ ] Import required libraries (pandas, sqlalchemy)
- [ ] Establish database connection with provided credentials
- [ ] Verify successful connection

### Step 2: Table Exploration
- [ ] Print the first rows of each table (`books`, `authors`, `publishers`, `ratings`, `reviews`)
- [ ] Document the structure and relationships between tables

### Step 3: SQL Queries
- [ ] **Query 1:** `SELECT COUNT(*) FROM books WHERE publication_date > '2000-01-01'`
- [ ] **Query 2:** Join `books` with `reviews` and `ratings` using `GROUP BY book_id` with `COUNT()` and `AVG()`
- [ ] **Query 3:** Join `books` with `publishers`, filter `num_pages > 50`, group by publisher and order by count DESC
- [ ] **Query 4:** Join `books`, `authors`, `ratings`; filter books with `COUNT(rating) >= 50`; get author with highest `AVG(rating)`
- [ ] **Query 5:** Subquery to identify users with >50 ratings in `ratings`, then calculate average reviews in `reviews`

### Step 4: Results Documentation
- [ ] Store each result in a pandas DataFrame
- [ ] Write interpretive conclusions for each query
- [ ] Connect findings with the business context (value proposition for the new product)
